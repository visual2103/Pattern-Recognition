**Rețelele Neuronale Convoluționale (CNNs)** sunt o clasă de rețele neuronale profunde utilizate în principal pentru analiza imaginilor. Acestea sunt folosite pe scară largă în sarcini precum clasificarea și segmentarea imaginilor, detectarea obiectelor și analiza video, datorită capacității lor de a învăța automat și adaptiv ierarhii de trăsături. 

Rețelele Neuronale Convoluționale (CNN) sunt conceptual similare cu Perceptronii Multistrat (MLP), fiind alcătuite din neuroni caracterizați prin ponderi și deplasamente ajustabile în timpul procesului de învățare. Fiecare neuron primește un set de intrări, realizează un produs scalar între acestea și ponderile asociate, iar ulterior, opțional, aplică o funcție de activare non-liniară pentru a introduce complexitate modelului. Întreaga rețea poate fi reprezentată ca o funcție diferențiabilă continuă, care mapază direct datele brute ale imaginii (pixelii) de la intrare către scorurile corespunzătoare claselor de la ieșire. De asemenea, rețelele CNN folosesc o funcție de pierdere în ultimul strat care este optimizată folosind un algoritm de optimizare iterativ precum gradient descent. Arhitecturile CNN presupun în mod explicit că datele de intrare sunt imagini, ceea ce permite integrarea directă a anumitor proprietăți specifice ale imaginilor în structura rețelei. Această abordare optimizează procesul de calcul al funcției de forward, făcându-l mai eficient, și reduce semnificativ numărul de parametri ai modelului, contribuind astfel la o utilizare mai eficientă a resurselor și la îmbunătățirea generalizării modelului.

Rețelele neuronale convoluționale se bazează pe mai multe concepte fundamentale care le permit să proceseze să proceseze imagini: 

1. **Stratul convoluțional**

Un strat de convoluție aplică operații de convoluție pe datele de intrare folosind un set de filtre. Aceste filtre se găsesc automat prin procesul de învățare. Scopul este să detecteze caracteristici locale, cum ar fi muchii, texturi sau modele complexe, pe baza valorilor pixelilor. Filtrul dintr-o rețea neuronală convoluțională este un volum 3D. Dimensiunile filtrului includ:

- Lățimea (de exemplu, 3 pentru un filtru 3x3)
- Înălțimea (de exemplu, 3 pentru un filtru 3x3)
- Adâncimea (sau numărul de canale) care este egală cu numărul de canale ale imaginii (sau volumului) de intrare.

În practică un strat de convoluție va învăța un set de filtre. Acestea se deplasează peste imagini cu un anumit pas denumit stride și se calculează produsul scalar între valorile filtrului și valorile corespunzătoare ale pixelilor din imagine. Rezultatul se înregistrează într-un element din harta de trăsături.

2. **Hartă de trăsături (Feature Maps/ Activations)**

Stratul de convoluție produce o hartă de trăsături pentru fiecare filtru aplicat. Fiecare hartă evidențiază zonele din imagine care se potrivesc trăsăturilor detectate de filtru. Dimensiunea ieșirii unui strat de convoluție este determinată de:

$$
W_{out} = \frac{W_{in}−F+2P}{S} + 1
$$

Unde:

- $W_{out}$ este dimensiunea ieșirii (lățime sau înălțime).
- $W_in$ este dimensiunea intrării (lățime sau înălțime).
- F este dimensiunea filtrului.
- P este dimensiunea padding-ului (padding-ul este procesul de adăugare a pixelilor (de obicei zero) în jurul marginilor volumului de intrare)
- S este stride (dimensiunea pasului de deplasare a convoluției)

3. **Funcția de activare**


După aplicarea convoluției, este necesară o funcție de activare pentru a introduce non-liniaritate în rețea. Funcția de activare permite modelului să învețe relații complexe între intrări și ieșiri.
Cea mai utilizată funcție de activare în CNN-uri este ReLU (Rectified Linear Unit):

$$
ReLU(x)=max(0,x)
$$

5. **Operația de pooling**

Pooling-ul este o operație utilizată în Rețelele Neuronale Convoluționale (CNN) pentru a reduce dimensiunile spațiale ale hărților de trăsături generate de straturile de convoluție. Scopul principal al pooling-ului este să diminueze dimensiunea datelor, păstrând în același timp caracteristicile esențiale, ceea ce face rețeaua mai eficientă și mai rezistentă la mici variații sau translații în datele de intrare. Pooling-ul acționează pe regiuni mici ale unei hărți de caracteristici și înlocuiește fiecare regiune cu o singură valoare, bazată pe o anumită funcție (de exemplu, maximul sau media valorilor).

Diminuarea dimensiunii spațiale: De exemplu, dacă harta de caracteristici are dimensiunea 32x32 și pooling-ul folosește un patch 2x2 cu stride 2, dimensiunea rezultată va fi redusă la 16x16. Tipurile cele mai folosite de pooling sunt:

- Max Pooling: Selectează valoarea maximă din fiecare regiune
- Average Pooling: Calculează media valorilor din fiecare regiune

5. **Straturi complet conectate (Fully Connected Layers)**

   
După mai multe straturi convoluționale și de pooling, ieșirea finală este transformată din matrice într-un vector. Vectorul este trecut prin unul sau mai multe straturi complet conectate, unde fiecare neuron este conectat la toți neuronii stratului anterior. Aceste straturi se folosesc pentru extragerea informației globale.

7. **Batch normalization**

Batch Normalization (BN) este un strat care normalizează activările din interiorul rețelei și este folosit pentru a accelera procesul de antrenare al rețelelor și pentru a îmbunătăți stabilitatea lor. Introducerea acestei tehnici a avut un impact semnificativ asupra performanței și antrenării modelelor adânci, reducând sensibilitatea la inițializarea greșită a parametrilor și la setările de hiperparametri. Stratul de Batch Normalization normalizează valorile intermediare (activările) ale unui strat, astfel încât să aibă o distribuție cu media zero și deviația standard unitară, înainte de a aplica funcția de activare.

Pentru fiecare mini-lot (batch) de date, valorile activărilor sunt normalizate:
$$
\mu = \frac{1}{m} \sum_{i=1}^{m} x_i
$$

$$
\sigma^2 = \frac{1}{m} \sum_{i=1}^{m} (x_i - \mu)^2
$$

$$
\hat{x}_i = \frac{x_i - \mu}{\sqrt{\sigma^2 + \epsilon}}
$$


Apoi valorile obținute se re-scalează și deplasează (γ,β) pentru a păstra capacitatea modelului de a învăța:

$$
y_i = \gamma \hat{x}_i + \beta
$$

7. **Stratul de dropout**

Pentru a preveni supraadaptarea (overfitting) se pot dezactiva (seta la 0) aleatoriu un subset de ponderi în timpul antrenării. Acest lucru forțează rețeaua să învețe trăsături mai robuste, descurajând dependența de ponderi specifice.

`torch.utils.data.Dataset` este o clasă abstractă în PyTorch care definește structura unui dataset. Atunci când se crează un dataset personalizat, trebuie să se moștenească această clasă și să se implementeze două metode esențiale:

- `__len__()`: Această metodă returnează dimensiunea dataset-ului (numărul total de exemple). Este utilizată pentru a obține lungimea dataset-ului prin len(dataset).
- `__getitem__()`: Această metodă returnează un eșantion specific (de exemplu, o imagine și eticheta sa), corespunzător unui anumit index. Este utilizată pentru a accesa datele astfel: dataset[i], unde i este indexul.

Mai jos, se prezintă un exemplu de dataset personalizat pentru a accesa imagini și etichete din setul CIFAR-10.

````python
import pickle
from torch.utils.data import Dataset

file = "data/cifar-10-batches-py/data_batch_1" 
with open(file, 'rb') as fin:
  cifar_data_batch_1 = pickle.load(fin, encoding='bytes')
# keys: b'batch_label', b'labels', b'data', b'filenames'

print(cifar_data_batch_1[b'batch_label'])
print(cifar_data_batch_1[b'labels'][:20])
print(cifar_data_batch_1[b'data'].shape)

class CIFAR10CustomDataset(Dataset):

    def __init__(self, images, labels, transform=None):
        self.images = images
        self.labels = labels
        self.transform = transform 

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, index):
        
        img, label = self.images[index], self.labels[index]
        img = img.reshape(3, 32, 32).transpose(1, 2, 0)

        if self.transform is not None:
            img = self.transform(img)

        return img, label

batch_size = 32
trainsetCustom = CIFAR10CustomDataset(cifar_data_batch_1[b'data'], cifar_data_batch_1[b'labels'], transform=transforms.ToTensor())
trainloaderCustomDataset = torch.utils.data.DataLoader(trainsetCustom, batch_size=batch_size, shuffle=True, num_workers=0)

dataiter = iter(trainloaderCustomDataset) 
images, labels = next(dataiter) 

imshow(torchvision.utils.make_grid(images))
print(' '.join(f'{classes[labels[j]]:5s}' for j in range(batch_size)))
````

## Definirea arhitecturii

În PyTorch, o rețea neuronală este definită ca o clasă Python. Aceasta trebuie să conțină două metode esențiale:

- `__init__()`: Inițializează straturile folosite în rețea.
- `forward()`: Definește modul în care rețeaua efectuează propagarea înainte (forward pass).

PyTorch generează automat o metodă `backward()`, care calculează gradienții pe baza operațiilor efectuate în metoda `forward()`.

Toate blocurile de bază pentru construirea rețelelor neuronale definite în PyTorch pot fi găsite în documentația [torch.nn](https://pytorch.org/docs/stable/nn.html).

Un exemplu de rețea MLP se regăsește mai jos:

````python
import torch.nn as nn

class MLP(nn.Module):
    def __init__(self):
        super().__init__()
        self.layers = nn.Sequential(
            nn.Flatten(),
            nn.Linear(32*32, 25),
            nn.ReLU(),
            nn.Linear(25, 10)
        )

    def forward(self, x):
        return self.layers(x)

model = MLP()
print(model)
````

**Ex4. Implementați următoarea architectura CNN** folosind straturile: `nn.Conv2d`, `nn.BatchNorm2d`, `nn.MaxPool2d`, `nn.Linear`.

1. **Input**:
   - CIFAR-10 images: $(3 \times 32 \times 32 )$ (RGB).

2. **Convolutional Block 1**:
   - Convolutional Layer: $(3 \times 3)$ filters, 32 filters, stride=1, padding=1.
   - Batch Normalization.
   - ReLU Activation.
   - MaxPooling: $(2 \times 2)$ pool size, stride=2.

3. **Convolutional Block 2**:
   - Convolutional Layer: $(3 \times 3)$ filters, 64 filters, stride=1, padding=1.
   - Batch Normalization.
   - ReLU Activation.
   - MaxPooling: $(2 \times 2)$ pool size, stride=2.

4. **Fully Connected Layers**:
   - Flatten the feature maps.
   - Linear Layer: 512 neurons with ReLU activation.
   - Linear Layer: 10 neurons (output layer, one for each CIFAR-10 class).

````python
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.conv1 = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )
        self.conv2 = nn.Sequential(
            nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )
        self.fc = nn.Sequential(
            nn.Flatten(),
            nn.Linear(64 * 8 * 8,512),
            nn.ReLU(),
            nn.Linear(512,10)
        )

    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.fc(x)
        return x

model = CNN()
print(model)
````

## Antrenarea rețelei

Pentru a antrena modelul, trebuie să definim o funcție de pierdere (loss function) și un optimizator. Pentru o sarcină de clasificare, utilizăm de obicei funcția de pierdere Cross-entropy. În PyTorch, putem folosi clasa [CrossEntropyLoss](https://pytorch.org/docs/stable/generated/torch.nn.CrossEntropyLoss.html).

Notă: În mod normal, după ultimul strat, aplicăm funcția softmax care transformă scorurile în probabiliăți normalizate. În PyTorch, funcția de pierdere `CrossEntropyLoss` așteaptă să primească scorurile brute ale modelului, deoarece funcția softmax este inclusă în implementarea `CrossEntropyLoss`. Din acest motiv, nu trebuie să folosim explicit softmax în definirea rețelei de mai sus.

$$
\ell(x, y) = \text{CrossEntropyLoss} = \{l_1, \dots, l_N\}^\top, \quad 
l_n = -\log \frac{\exp(x_{n, y_n})}{\sum_{c=1}^C \exp(x_{n, c})}
$$

- Numărătorul: $\exp(x_{n, y_n})$ calculează probabilitatea nenormalizată pentru clasa corectă $y_n$ pentru exemplul n.
- Numitorul: $\sum_{c=1}^C \exp(x_{n, c})$ calculează suma probabilităților nenormalizate pentru toate cele C clase.

În final, trebuie să definim un optimizator, care specifică modul în care se actualizează parametrii modelului pe baza gradienților calculați. PyTorch implementează mai mulți algoritmi diferiți pentru optimizare în modulul [nn.optim](https://pytorch.org/docs/stable/optim.html): Stochastic Gradient Descent, Adam, RMSprop etc.

````python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(params=model.parameters(), lr=0.01)
````

În PyTorch scriem manual bucla de antrenare.

Codul de mai jos constă din două bucle:

1. **Bucla exterioară**:
   - Trecem printr-un număr de epoci. O **epocă** reprezintă o parcurgere completă a întregului set de antrenare.

2. **Bucla interioară**:
   - Parcurge întregul set de date în iterații, procesând câte un lot (batch) la fiecare iterație.

---

Pentru fiecare lot, facem următorii pași:

1. **Propagarea înainte (Forward Pass)**:
   - Efectuăm propagarea înainte prin model:  `output = model(data)`
    
    
2. **Calculul funcției de pierdere**:
   - Calculăm pierderea (**loss**) - adică eroarea dintre ieșirea rețelei și ținta dorită - folosind funcția de pierdere definită anterior: `loss = criterion(output, target)`

3. **Propagarea înapoi și actualizarea ponderilor**:
   - Efectuăm propagarea înapoi pentru a calcula gradienții: `loss.backward()`
   - Actualizăm ponderile modelului folosind optimizatorul: `optimizer.step()`
   - Resetăm gradienții la zero pentru următoarea iterație: `optimizer.zero_grad()`

````python
from torch.utils.tensorboard import SummaryWriter

writer = SummaryWriter("runs/experiment1")
def train(data_loader, model, criterion, optimizer, epoch):
    model.train()

    num_batches = len(data_loader)
    num_items = len(data_loader.dataset)

    total_loss = 0
    total_correct = 0
    for data, target in data_loader:
        # Do a forward pass
        output = model(data)
        #print(output)
        # Calculate the loss
        loss = criterion(output, target)
        total_loss += loss.item()
        _, predicted = torch.max(output.data, 1)
        total_correct += (predicted == target).sum().item()
          
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()

    train_loss = total_loss/num_batches
    accuracy = total_correct/num_items
    print(f"Average loss: {train_loss:7f}, Accuracy: {accuracy:.7f}")

    print(f"Average loss: {train_loss:7f}")
    writer.add_scalar("Loss/train", train_loss, epoch)
    writer.add_scalar("Accuracy/train", accuracy, epoch)
    return train_loss, accuracy 
````

````python
epochs = 20
model = CNN()

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(params=model.parameters(), lr=0.01)

for epoch in range(epochs):
    print(f"Training epoch: {epoch+1}")
    train(trainloaderCustomDataset, model, criterion, optimizer, epoch)
````

> Se va trece modelul din modul `train()` în modul `eval()` prin apelarea `model.eval()`

````python
### YOUR CODE HERE
def eval (data_loader, model):
    model.eval()
    num_batches = len(data_loader)
    num_items = len(data_loader.dataset)
    total_correct = 0
    for data, target in data_loader:
         output = model(data)
         _, predicted = torch.max(output, dim=1) 
         total_correct += (predicted == target).sum().item() 
    accuracy = total_correct / num_items
    print(f"Test Accuracy: {accuracy*100:.2f}%")

file = "data/cifar-10-batches-py/test_batch"
with open(file, 'rb') as fin:
    test_batch = pickle.load(fin, encoding='bytes')

testsetCustom = CIFAR10CustomDataset(test_batch[b'data'], test_batch[b'labels'], transform=transforms.ToTensor())
testloaderCustom = torch.utils.data.DataLoader(testsetCustom, batch_size=32, shuffle=False)
eval(testloaderCustom, model)
````
